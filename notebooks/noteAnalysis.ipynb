```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../data/database_saopaulo.csv')

df['Data'] = pd.to_datetime(df['Data'], format="%d/%m/%Y")

def playedOscar(df):
    if df['Data'] >= pd.Timestamp('2025-01-23'):
        return 'Sim'
    else:
        return 'Não'
df['Oscar Jogou?'] = df.apply(playedOscar, axis=1)

df['Passes Totais (Oscar)'] = df['Passes Totais (Oscar)'].fillna(0)
df['Passes Certos (Oscar)'] = df['Passes Certos (Oscar)'].fillna(0)
df['Passes Decisivos (Oscar)'] = df['Passes Decisivos (Oscar)'].fillna(0)
df['Gols (Oscar)'] = df['Gols (Oscar)'].fillna(0)
df['Assistências (Oscar)'] = df['Assistências (Oscar)'].fillna(0)

df['Passes Totais (Oscar)'] = df['Passes Totais (Oscar)'].astype(int)
df['Passes Certos (Oscar)'] = df['Passes Certos (Oscar)'].astype(int)
df['Passes Decisivos (Oscar)'] = df['Passes Decisivos (Oscar)'].astype(int)
df['Gols (Oscar)'] = df['Gols (Oscar)'].astype(int)
df['Assistências (Oscar)'] = df['Assistências (Oscar)'].astype(int)

df['Posse (%)'] = df['Posse (%)'].str.replace('%', '', regex=False)
df['Posse (%)'] = df['Posse (%)'].astype(float)

df['Pontuação SofaScore (Oscar)'] = df['Pontuação SofaScore (Oscar)'].str.replace(',', '.', regex=False)
df['Pontuação SofaScore (Oscar)'] = df['Pontuação SofaScore (Oscar)'].astype(float)
df['Pontuação SofaScore (Oscar)'] = df['Pontuação SofaScore (Oscar)'].fillna(0)

df['Passes Certos (%)'] = ((df['Passes Certos'] * 100)/df['Passes Totais']).astype(float).round(1)
df['Passes Certos (Oscar) (%)'] = ((df['Passes Certos (Oscar)'] * 100)/df['Passes Totais (Oscar)']).astype(float).round(1)
df['Passes Certos (Oscar) (%)'] = df['Passes Certos (Oscar) (%)'].fillna(0)
df['Passes Certos (SP X Oscar) (%)'] = ((df['Passes Certos (Oscar)'] * 100)/df['Passes Certos']).astype(float).round(1)
df['Participações Decisivas (Oscar)'] = df[['Passes Decisivos (Oscar)', 'Gols (Oscar)', 'Assistências (Oscar)']].sum(axis=1).astype(int)
df['Aproveitamento (Oscar) (%)'] = (((df['Gols (Oscar)'] + df['Assistências (Oscar)']) * 100)/df['Participações Decisivas (Oscar)']).astype(float).round(1)
df['Aproveitamento (Oscar) (%)'] = df['Aproveitamento (Oscar) (%)'].fillna(0)

df = df.rename(columns={
    'Chances Criadas (SPFC)': 'Chances Criadas'
})

mediaPasseOscar = df[df['Oscar Jogou?'] == 'Sim']['Passes Certos (%)'].mean().round(1)
mediaPasse = df[df['Oscar Jogou?'] == 'Não']['Passes Certos (%)'].mean().round(1)
mediaPassesOscar = df[df['Oscar Jogou?'] == 'Sim']['Passes Certos (SP X Oscar) (%)'].mean().round(1)
mediaPassesCertosOscar = df[df['Oscar Jogou?'] == 'Sim']['Passes Certos (Oscar) (%)'].mean().round(1)

mediaChancesOscar = df[df['Oscar Jogou?'] == 'Sim']['Chances Criadas'].mean().round(1)
mediaChances = df[df['Oscar Jogou?'] == 'Não']['Chances Criadas'].mean().round(1)

mediaChutesOscar = df[df['Oscar Jogou?'] == 'Sim']['Chutes'].mean().round(1)
mediaChutes = df[df['Oscar Jogou?'] == 'Não']['Chutes'].mean().round(1)

mediaParticipações = df[df['Oscar Jogou?'] == 'Sim']['Participações Decisivas (Oscar)'].mean().round(1)

mediaAproveitamento = df[df['Oscar Jogou?'] == 'Sim']['Aproveitamento (Oscar) (%)'].mean().round(1)

mediaPontuação = df[df['Oscar Jogou?'] == 'Sim']['Pontuação SofaScore (Oscar)'].mean().round(1)

print('=== ANÁLISE: ==========================================================================================\n')
# print(f'Média de passes certos SEM Oscar: {mediaPasse}%')
# print(f'Média de passes certos COM Oscar: {mediaPasseOscar}%')
print(f'Oscar participou, em média, de {mediaPassesOscar}% dos passes certos do time com aproveitamento médio de {mediaPassesCertosOscar}%')

print(f'\nMédia de chances criadas SEM Oscar: {mediaChances}')
print(f'Média de chances criadas COM Oscar: {mediaChancesOscar}')

print(f'\nMédia de chutes SEM Oscar: {mediaChutes}')
print(f'Média de chutes COM Oscar: {mediaChutesOscar}')

print(f'\nOscar participou em média de {mediaParticipações} ações decisivas, com aproveitamento (resultando em gols) de {mediaAproveitamento}%')

print(f'\nPontuação SofaScore média: {mediaPontuação}')

# TODO:
# print(f'Melhor jogo do Oscar foi contra o {time}, com pontuação {pontuação}, participando de {AssGols} gols e {açõesDecisivas} ações decisivas'
print('\n=======================================================================================================')

df
```
